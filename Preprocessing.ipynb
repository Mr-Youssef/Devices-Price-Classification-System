# Data Preprocessing

### Loading the Data

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import pickle

df = pd.read_excel("Dataset/train.xlsx")

print("Number of records: ", len(df))
```

```python
df.head()
```

```python
df.info()
```

### Observation:
#### Notice that all features are numerical features which is easier to deal with.

### Handling Missing Values

```python
print("Total number of null values: ", df.isnull().sum().sum())
print("\nCount of null values per feature: ")
print(df.isnull().sum())
```

### Observation:
#### The biggest number of missing values in a feature is 5 

### Conclusion: 

#### Since number of nulls is minimal, we can safely drop them

```python
# Drop the rows with null values
df = df.dropna()
print("Number of records after dropping nulls: ", len(df))
```

### Handling Duplicates

```python
# Checking for duplicates
df.duplicated().sum()
# No duplicates found
```

### Showing some statistics

```python
df.describe()
```

### Observation
#### Notice that the features have different ranges. Therefore, we need to normalize them.

### Features Normalization

```python
# Normalizing the data except the last column (price_range which is the target)
scaler = MinMaxScaler()
df.iloc[:, :-1] = scaler.fit_transform(df.iloc[:, :-1])
df.head()       # Checking the data after normalization
```

```python
# Saving the scaler for later use
pickle.dump(scaler, open('GeneratedFiles/scaler.pkl', "wb"))
```

```python
# Checking the distribution of the target
df.value_counts('price_range')
```

### Observation:
#### The dataset is balanced as the distribution of the target values is almost equal

### Saving the data

```python
df.to_excel('Dataset/PreprocessedTrain.xlsx', index=False)
```
